```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
size = 300
X = np.random.rand(size)*5-2.5
w4, w3, w2, w1, w0 = 1, 2, 1, -4, 2
y = w4*(X**4) + w3*(X**3) + w2*(X**2) + w1*X + w0 + np.random.randn(size)*8-4
df = pd.DataFrame({'x': X, 'y': y})
df.to_csv('dane_do_regresji.csv',index=None)
df.plot.scatter(x='x',y='y')
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[['x']], df[['y']], test_size = 0.20, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg1 = LinearRegression()
lin_reg1.fit(X_train, y_train)
plt.scatter(X_train,y_train,color='g')
plt.scatter(X_train, lin_reg1.predict(X_train),color='k')
```

```python
lin_reg2 = LinearRegression()
lin_reg2.fit(X_test,y_test)
plt.scatter(X_test, y_test, color = 'b')
plt.scatter(X_test, lin_reg2.predict(X_test), color='r')
```

```python
import sklearn.neighbors
knn3_reg = sklearn.neighbors.KNeighborsRegressor(n_neighbors=3)
knn3_reg.fit(X_train, y_train)
plt.scatter(X_train, y_train, color = 'b')
plt.scatter(X_train, knn3_reg.predict(X_train), color='r')
```

```python
import sklearn.neighbors
knn3t_reg = sklearn.neighbors.KNeighborsRegressor(n_neighbors=3)
knn3t_reg.fit(X_test, y_test)
plt.scatter(X_test, y_test, color = 'b')
plt.scatter(X_test, knn3t_reg.predict(X_test), color='r')
```

```python
knn5_reg = sklearn.neighbors.KNeighborsRegressor(n_neighbors=5)
knn5_reg.fit(X_train, y_train)
plt.scatter(X_train, y_train, color = 'b')
plt.scatter(X_train, knn5_reg.predict(X_train), color='r')
```

```python
knn5t_reg = sklearn.neighbors.KNeighborsRegressor(n_neighbors=5)
knn5t_reg.fit(X_test, y_test)
plt.scatter(X_test, y_test, color = 'b')
plt.scatter(X_test, knn5t_reg.predict(X_test), color='r')
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_features2 = PolynomialFeatures(degree=2, include_bias=False)
poly_2_reg = poly_features2.fit_transform(X_train)
lin_reg_p2 = LinearRegression()
lin_reg_p2.fit(poly_2_reg, y_train)
plt.scatter(X_train , y_train, color = 'b')
plt.plot(X_train , lin_reg_p2.predict( poly_features2.fit_transform(X_train)), color = 'r')
plt.show()

#testowy
poly_features2t = PolynomialFeatures(degree=2, include_bias=False)
poly_2t_reg = poly_features2t.fit_transform(X_test)
lin_reg_p2t = LinearRegression()
lin_reg_p2t.fit(poly_2t_reg, y_test)
plt.scatter(X_test , y_test, color = 'b')
plt.plot(X_test , lin_reg_p2t.predict( poly_features2t.fit_transform(X_test)), color = 'r')
plt.show()
```

```python
poly_features3 = PolynomialFeatures(degree=3, include_bias=False)
poly_3_reg = poly_features3.fit_transform(X_train)
lin_reg_p3 = LinearRegression()
lin_reg_p3.fit(poly_3_reg, y_train)
plt.scatter(X_train , y_train, color = 'b')
plt.plot(X_train , lin_reg_p3.predict( poly_features3.fit_transform(X_train)), color = 'r')
plt.show()

#testowy
poly_features3t = PolynomialFeatures(degree=3, include_bias=False)
poly_3t_reg = poly_features3t.fit_transform(X_test)
lin_reg_p3t = LinearRegression()
lin_reg_p3t.fit(poly_3t_reg, y_test)
plt.scatter(X_test , y_test, color = 'b')
plt.plot(X_test , lin_reg_p3t.predict( poly_features3t.fit_transform(X_test)), color = 'r')
plt.show()
```

```python
poly_features4 = PolynomialFeatures(degree=4, include_bias=False)
poly_4_reg = poly_features4.fit_transform(X_train)
lin_reg_p4 = LinearRegression()
lin_reg_p4.fit(poly_4_reg, y_train)
plt.scatter(X_train , y_train, color = 'b')
plt.plot(X_train , lin_reg_p4.predict( poly_features4.fit_transform(X_train)), color = 'r')
plt.show()

#testowy
poly_features4t = PolynomialFeatures(degree=4, include_bias=False)
poly_4t_reg = poly_features4t.fit_transform(X_test)
lin_reg_p4t = LinearRegression()
lin_reg_p4t.fit(poly_4t_reg, y_test)
plt.scatter(X_test , y_test, color = 'b')
plt.plot(X_test , lin_reg_p4t.predict( poly_features4t.fit_transform(X_test)), color = 'r')
plt.show()
```

```python
poly_features5 = PolynomialFeatures(degree=5, include_bias=False)
poly_5_reg = poly_features5.fit_transform(X_train)
lin_reg_p5 = LinearRegression()
lin_reg_p5.fit(poly_5_reg, y_train)
plt.scatter(X_train , y_train, color = 'b')
plt.plot(X_train , lin_reg_p5.predict( poly_features5.fit_transform(X_train)), color = 'r')
plt.show()

#testowy
poly_features5t = PolynomialFeatures(degree=5, include_bias=False)
poly_5t_reg = poly_features5t.fit_transform(X_test)
lin_reg_p5t = LinearRegression()
lin_reg_p5t.fit(poly_5t_reg, y_test)
plt.scatter(X_test , y_test, color = 'b')
plt.plot(X_test , lin_reg_p5t.predict( poly_features5t.fit_transform(X_test)), color = 'r')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
a=mean_squared_error(y_train, lin_reg1.predict(X_train))
b=mean_squared_error(y_test, lin_reg2.predict(X_test))

mse = pd.DataFrame(np.array([[a, b,'lin_reg']]),columns=['train_mse','test_mse','nazwa'])
mse=mse.set_index('nazwa')
```
