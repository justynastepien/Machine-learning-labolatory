```python
import tensorflow_datasets as tfds
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import keras
from functools import partial
import pickle
```

```python
[test_set_raw, valid_set_raw, train_set_raw], info = tfds.load(
    "tf_flowers",
    split=["train[:10%]", "train[10%:25%]", "train[25%:]"],
    as_supervised = True,
    with_info = True)
```

```python
class_names = info.features["label"].names
n_classes = info.features["label"].num_classes
dataset_size = info.splits["train"].num_examples
```

```python
plt.figure(figsize=(12,8))
index = 0
sample_images = train_set_raw.take(9)
for image, label in sample_images:
    index += 1
    plt.subplot(3, 3, index)
    plt.imshow(image)
    plt.title("Class: {}".format(class_names[label]))
    plt.axis("off")
    
plt.show(block=False)
```

```python
def preprocess(image, label):
    resized_image = tf.image.resize(image, [224, 224])
    return resized_image, label
```

```python
batch_size = 32
train_set = train_set_raw.map(preprocess).shuffle(dataset_size).batch(batch_size).prefetch(1)
valid_set = valid_set_raw.map(preprocess).batch(batch_size).prefetch(1)
test_set = test_set_raw.map(preprocess).batch(batch_size).prefetch(1)
```

```python
plt.figure(figsize = (8, 8))
sample_batch = train_set.take(1)
for X_batch, y_batch in sample_batch:
    for index in range(12):
        plt.subplot(3, 4, index + 1)
        plt.imshow(X_batch[index]/255.0)
        plt.title("Class: {}".format(class_names[y_batch[index]]))
        plt.axis("off")

plt.show()
```

```python
DefaultConv2D=partial(keras.layers.Conv2D, kernel_size=3, activation='relu', padding="SAME")

model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Rescaling(scale = 1/127.5, offset=-1))
model.add(DefaultConv2D(filters=64, kernel_size=7, input_shape=[28,28,1]))
model.add(keras.layers.MaxPooling2D(pool_size=2))
model.add(DefaultConv2D(filters=128))
model.add(DefaultConv2D(filters=128))
model.add(keras.layers.MaxPooling2D(pool_size=2))
model.add(DefaultConv2D(filters=256))
model.add(DefaultConv2D(filters=256))
model.add(keras.layers.MaxPooling2D(pool_size=2))
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(units=128, activation='relu'))
model.add(keras.layers.Dropout(0.5))
model.add(keras.layers.Dense(units=10, activation='softmax'))
```

```python
model.compile(loss = "sparse_categorical_crossentropy", optimizer = "sgd", metrics = ['accuracy'])
```

```python
model.fit(train_set, validation_data=valid_set, epochs = 10)
```

```python
acc_train = model.history.history.get('accuracy')
acc_train.reverse()
print(acc_train[0])
```

```python
acc_test = model.evaluate(test_set)
print[acc_test[1]]
```

```python
acc_valid = model.history.history.get('val_accuracy')
acc_valid.reverse()
print(acc_valid[0])

#(acc_train, acc_valid, acc_test)
```

```python
krot1 = (acc_train[0], acc_valid[0], acc_test[1])
f1 = open("simple_cnn_acc.pkl", "wb")
pickle.dump(krot1, f1)
```

```python
def preprocess(image, label):
    resized_image = tf.image.resize(image, [224, 224])
    final_image = tf.keras.applications.xception.preprocess_input(resized_image)
    return final_image, label
```

```python
plt.figure(figsize=(8, 8))
sample_batch = train_set.take(1)
for X_batch, y_batch in sample_batch:
    for index in range(12):
        plt.subplot(3, 4, index + 1)
        plt.imshow(X_batch[index] / 2 + 0.5)
        plt.title("Class: {}".format(class_names[y_batch[index]]))
        plt.axis("off")
plt.show()
```

```python
base_model = tf.keras.applications.xception.Xception(
weights="imagenet",
include_top=False)
```

```python
for index, layer in enumerate(base_model.layers):
    print(index, layer.name)
```

```python
avg = tf.keras.layers.GlobalAveragePooling2D()(base_model.output)
output = tf.keras.layers.Dense(units=128, activation='softmax')(avg)
```

```python
modelv2 = tf.keras.Model(inputs=base_model.input, outputs=output)
optimizer = tf.keras.optimizers.SGD(learning_rate=0.2, momentum=0.9,decay=0.01)
modelv2.compile(loss = "sparse_categorical_crossentropy", optimizer = optimizer, metrics = ['accuracy'])
```

```python
for layer in base_model.layers:
    layer.trainable = False
```

```python
history = modelv2.fit(train_set, validation_data=valid_set, epochs = 5)
```

```python
for layer in base_model.layers:
    layer.trainable = True
```

```python
modelv2.fit(train_set, validation_data=valid_set, epochs = 10)
```

```python
#lista1 = modelv2.evaluate(test_set)

acc_train = modelv2.history.history.get('accuracy')
acc_train.reverse()
print(acc_train[0])
#(acc_train, acc_valid, acc_test)
```

```python
acc_test = modelv2.evaluate(test_set)
acc_test[1]
```

```python
acc_valid = modelv2.history.history.get('val_accuracy')
acc_valid.reverse()
print(acc_valid[0])
```

```python
krot2 = (acc_train[0], acc_valid[0], acc_test[1])
f2 = open("xception_acc.pkl", "wb")
pickle.dump(krot2, f2)
```
